```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
file_path = 'data/bsmax_0_1.csv'
data = pd.read_csv(file_path, sep='\t')

data = data[:int(len(data)*0.8)]


# Separate features and target variable
X = data.drop('rtn', axis=1)
y = data['rtn']

# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
r2 = r2_score(y_test, y_pred)
print(f'R-squared value: {r2}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
file_path = 'data/bsmax_0_1.csv'
data = pd.read_csv(file_path, sep='\t')

# Split the data into 80% for training and 20% for prediction
split_index = int(len(data) * 0.8)
train_data = data[:split_index]
predict_data = data[split_index:]

# Separate features and target variable for training data
X_train = train_data.drop('rtn', axis=1)
y_train = train_data['rtn']

# Normalize the features for training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Train a linear regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Separate features and target variable for prediction data
X_predict = predict_data.drop('rtn', axis=1)
y_actual = predict_data['rtn']

# Normalize the features for prediction data
X_predict_scaled = scaler.transform(X_predict)

# Predict on the prediction set
y_pred = model.predict(X_predict_scaled)

# Evaluate the model on the prediction set
r2 = r2_score(y_actual, y_pred)
print(f'R-squared value on prediction data: {r2}')

# Visual comparison
plt.figure(figsize=(10, 6))
plt.plot(y_actual.values, label='Actual Values', marker='o')
plt.plot(y_pred, label='Predicted Values', marker='x')
plt.title('Actual vs Predicted Values')
plt.xlabel('Index')
plt.ylabel('Return')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# 데이터 로드
file_path = 'data/bsmax_0_1.csv'
data = pd.read_csv(file_path, sep='\t')

# 데이터 분할: 첫 80% 훈련, 마지막 20% 예측
split_index = int(len(data) * 0.8)
train_data = data[:split_index]
predict_data = data[split_index:]

# 훈련 데이터: 특징과 목표 변수 분리
X_train = train_data.drop('rtn', axis=1)
y_train = train_data['rtn']

# 특징 데이터 표준화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# 예측 데이터: 특징과 목표 변수 분리
X_predict = predict_data.drop('rtn', axis=1)
y_actual = predict_data['rtn']

# 예측 데이터 표준화
X_predict_scaled = scaler.transform(X_predict)

# 딥러닝 모델 정의
model = Sequential()
model.add(Dense(256, input_dim=X_train_scaled.shape[1], activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1))

# 모델 컴파일
model.compile(optimizer='adam', loss='mse')

# 조기 종료 콜백 설정
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# 모델 훈련
history = model.fit(X_train_scaled, y_train, epochs=300, batch_size=32, verbose=1, validation_split=0.2, callbacks=[early_stopping])

# 예측 수행
y_pred = model.predict(X_predict_scaled).flatten()

# 모델 평가
r2 = r2_score(y_actual, y_pred)
print(f'R-squared value on prediction data: {r2}')

# 시각적 비교
plt.figure(figsize=(10, 6))
plt.plot(y_actual.values, label='Actual Values', marker='o')
plt.plot(y_pred, label='Predicted Values', marker='x')
plt.title('Actual vs Predicted Values')
plt.xlabel('Index')
plt.ylabel('Return')
plt.legend()
plt.show()
```

머신러닝 vs 딥러닝

데이터 크기:

머신러닝: 수천에서 수십만 개의 데이터 포인트로도 충분히 잘 작동합니다.
딥러닝: 일반적으로 수십만 개 이상의 대규모 데이터셋이 필요합니다.

데이터 복잡성:

머신러닝: 상대적으로 단순한 데이터셋에서 좋은 성능을 보입니다.
딥러닝: 비정형 데이터(이미지, 텍스트 등)와 같은 복잡한 패턴을 더 잘 학습합니다.

모델 해석 가능성:

머신러닝: 모델이 비교적 해석 가능하고, 중요한 피처를 쉽게 파악할 수 있습니다.
딥러닝: 모델이 블랙박스에 가까워 해석이 어려운 경우가 많습니다.

컴퓨팅 자원:

머신러닝: 상대적으로 적은 컴퓨팅 자원으로도 잘 작동합니다.
딥러닝: GPU와 같은 고성능 하드웨어가 필요합니다.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import cross_val_score, KFold

# 데이터 로드
file_path = 'data/bsmax_0_1.csv'
data = pd.read_csv(file_path, sep='\t')

# 데이터 분할
X = data.drop('rtn', axis=1)
y = data['rtn']

# 특징 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 모델 리스트
models = {
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42)
}

# 교차 검증 설정
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 모델 평가 및 비교
for name, model in models.items():
    scores = cross_val_score(model, X_scaled, y, cv=kf, scoring='r2')
    print(f'{name} R-squared values: {scores}')
    print(f'{name} Mean R-squared value: {np.mean(scores)}')
```
